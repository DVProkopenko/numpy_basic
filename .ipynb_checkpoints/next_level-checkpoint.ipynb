```python
import numpy as np
p = np.poly1d([3, 2, -1])
p(0)
```

```python
p.roots
```

```python
p.order
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 20)
y = np.cos(x) + 0.3*np.random.rand(20)
p = np.poly1d(np.polyfit(x, y, 3))
```

```python
t = np.linspace(0, 1, 200)
plt.plot(x, y, 'o', t, p(t), '-')   
```

NumPy также имеет более сложный полиномиальный интерфейс, который поддерживает, например, Чебышева.

```python
p = np.polynomial.Polynomial([-1, 2, 3]) # coefs in different order!
p(0)
```

```python
p.roots()
```

```python
p.degree()  # In general polynomials do not always expose 'order'
```

Пример использования полиномов в базисе Чебышева для полиномов в диапазоне [-1, 1]: 

```python
x = np.linspace(-1, 1, 2000)
y = np.cos(x) + 0.3*np.random.rand(2000)
p = np.polynomial.Chebyshev.fit(x, y, 90)
```

```python
t = np.linspace(-1, 1, 200)
plt.plot(x, y, 'r.')   
```

```python
plt.plot(t, p(t), 'k-', lw=3)   
```

Полиномы Чебышева имеют некоторые преимущества в интерполяции.
